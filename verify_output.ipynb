```python
OUTPUT_DIR="output_experimental"
OUTPUT_TIME="time"

EXECUTIONS = 50
CITIES = [13, 14, 15]
THREADS = [1, 2, 3, 6, 9, 12]
ITERATIONS = 10
```

```python
def extract_results_from_file(file_name):
    file = open(file_name)
    data = []
    for exec in range(1, EXECUTIONS + 1):
        data.append(float(file.readline().split(" ")[0]))
    file.close()
    return data
```

```python
for city in list(CITIES):
    sequential_output_value = None
    for iteration in range(1,ITERATIONS + 1):
        file_name = OUTPUT_DIR + "/" + str(iteration) + "_seq_" + str(city) + "_cities_" + str(EXECUTIONS) + "_executions_result.out"
        
        instance_array = extract_results_from_file(file_name)
        
        if (iteration == 1):
            sequential_output_value = instance_array

            
    for thread in list(THREADS):
        for iteration in range(1,ITERATIONS + 1):
            file_name = OUTPUT_DIR + "/" + str(iteration) + "_par_" + str(city) + "_cities_" + str(EXECUTIONS) + "_executions_" + str(thread) + "_threads_result.out"
            
            parallel_output_value = extract_results_from_file(file_name)

            if (sequential_output_value == parallel_output_value):
                print("OK Cities: {} - Thread: {} - Iteration: {}".format(city, thread, iteration))
            else:
                print("ERROR Cities: {} - Thread: {} - Iteration: {}".format(city, thread, iteration))

```
